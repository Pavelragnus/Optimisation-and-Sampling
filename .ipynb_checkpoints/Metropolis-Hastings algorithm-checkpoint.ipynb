# Problem 6 

Write a program that generates values from normal distribution 
 using rejection sampling. It is necessary to generate a sample using random samples from the Cauchy distribution which in turn is generated using inverse CDF function. So, only a standard uniform distribution generator can be used as a source of randomness. All the distributions’ parameters must be kept as funciton parameters.

```python
import numpy as np
import matplotlib.pyplot as plt

def standard_uniform(size=1):
    return np.random.rand(size)

def cauchy_inverse_cdf(u, location=0, scale=1):
    return location + scale * np.tan(np.pi * (u - 0.5))

def normal_rejection_sampling(n, mu=0, sigma=1, cauchy_location=0, cauchy_scale=1):
    samples = []
    
    while len(samples) < n:
        u1, u2 = standard_uniform(2)
        x = cauchy_inverse_cdf(u1, location=cauchy_location, scale=cauchy_scale)
        acceptance_prob = np.exp(-0.5 * ((x - mu) / sigma)**2) / (np.sqrt(2 * np.pi) * sigma)
        
        if u2 < acceptance_prob:
            samples.append(x)
    
    return np.array(samples)

# Example usage
num_samples = 1000
generated_samples = normal_rejection_sampling(num_samples)

# Plot the histogram of generated samples
plt.hist(generated_samples, bins=50, density=True, alpha=0.7, color='blue', label='Generated Samples')
plt.title('Rejection Sampling: Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

# Problem 7

1.Write a program that generates a sample from multivariate normal distribution 
. It is necessary to use Metropolis-Hastings algorithm (see section “Metropolis algorithm (symmetric proposal distribution)” to generate a sample.   
2. As a proposing distribution a uniform distribution on a unit circular disk centered at current point in MCMC should be used. Uniform point inside unit disk can be generated using sample of size two from standard uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

def multivariate_normal(x, mean, cov):
    """
    Calculate the probability density of a multivariate normal distribution.
    """
    k = len(mean)
    det_cov = np.linalg.det(cov)
    inv_cov = np.linalg.inv(cov)
    exponent = -0.5 * (x - mean).T @ inv_cov @ (x - mean)
    return (1.0 / ((2 * np.pi) ** (k / 2) * det_cov ** 0.5)) * np.exp(exponent)

def metropolis_hastings(initial_sample, num_samples, mean, cov):
    """
    Generate samples from a multivariate normal distribution using Metropolis-Hastings algorithm.
    """
    current_sample = initial_sample
    samples = [current_sample]

    for _ in range(num_samples):
        # Propose a new sample from a uniform distribution on a unit circular disk
        proposal = current_sample + np.random.uniform(-1, 1, size=current_sample.shape)
        while np.linalg.norm(proposal) > 1:
            proposal = current_sample + np.random.uniform(-1, 1, size=current_sample.shape)

        # Calculate acceptance ratio
        alpha = min(1, multivariate_normal(proposal, mean, cov) /
                       multivariate_normal(current_sample, mean, cov))

        # Accept or reject the proposed sample
        if np.random.rand() < alpha:
            current_sample = proposal

        samples.append(current_sample)

    return np.array(samples)

# Parameters for the multivariate normal distribution
mean = np.array([0, 0])
covariance = np.array([[1, 0.5], [0.5, 1]])

# Initial sample
initial_sample = np.array([0, 0])

# Number of samples to generate
num_samples = 1000

# Generate samples using Metropolis-Hastings algorithm
samples = metropolis_hastings(initial_sample, num_samples, mean, covariance)

# Plot the generated samples
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.5)
plt.title('Metropolis-Hastings Samples from Multivariate Normal Distribution')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
