Derivatives of $x^2+y^4+z^6$: $2x$, $4y^3$, $6z^5$   
So at the initial point (1,1,1) the newton's direction is $-df(1,1,1) = (-2,-4,-6)$

### another

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f(x, y, z):
    return x**2 + y**4 + z**6

def ternary_search_3d(initial_point, newton_direction, max_iterations=100, step_size=0.01):
    trajectory = [initial_point]
    current_point = np.array(initial_point)
    
    for iteration in range(max_iterations):
        x, y, z = current_point
        fx = f(x, y, z)  # Calculate function value at the current point

        # Evaluate function values at two points along the Newton's direction
        fx1 = f(x, y, z)
        fx2 = f(x + step_size * newton_direction[0], 
                y + step_size * newton_direction[1], 
                z + step_size * newton_direction[2])
        
        # Update the current point based on the ternary search
        if fx2 < fx1:
            current_point = current_point + step_size * newton_direction
        else:
            current_point = current_point - step_size * newton_direction

        trajectory.append(tuple(current_point))

    return np.array(trajectory)

# Initial point and Newton's direction
initial_point = np.array((1, 1, 1))
newton_direction = np.array((-2, -4, -6))

# Apply ternary search
trajectory = ternary_search_3d(initial_point, newton_direction)

# Plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plotting the function surface as a 3D scatter plot
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
z = np.linspace(-2, 2, 100)
X, Y, Z = np.meshgrid(x, y, z, indexing='ij')
F = X**2 + Y**4 + Z**6
ax.scatter(X, Y, Z, c=F, cmap='winter_r', marker='.', alpha=0.002)


# Plotting the trajectory of the algorithm using a scatter plot and lines
ax.scatter(trajectory[:, 0], trajectory[:, 1], trajectory[:, 2], c='red', marker='o', label='Algorithm Trajectory', s=10)
#ax.plot(trajectory[:, 0], trajectory[:, 1], trajectory[:, 2], c='red', linewidth=2)

# Highlighting the initial and final points
#ax.scatter(initial_point[0], initial_point[1], initial_point[2], c='g', marker='o', label='Initial Point')
#ax.scatter(trajectory[-1, 0], trajectory[-1, 1], trajectory[-1, 2], c='b', marker='o', label='Final Point')



ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Ternary Search Trajectory in 3D Space')
ax.legend()

plt.show()
```

# Knapsack problem

## Part 0 

Given a set of items of size n with their prices p and weights w find a subset with the maximum total prize and the total weight not exceeding a predefined value W. Using python-mip or any other MIP solver write a small program that solves the knapsack problem for a set of items taken from an input file(one item per line, item’s name prize weight).

Here is an example input file:   
Laptop  10  0.5   
Flashcard   5   0.01  
Cryptokey 1 0.01   
Lamp 1 5.0   
Battery 1 0.05   
Charger 0 0.01   

```python
from pulp import LpProblem, LpVariable, lpSum, LpMaximize

def knapsack_solver(input_file, max_weight):
    items = []

    # Read input file
    with open(input_file, 'r') as file:
        for line in file:
            name, prize, weight = line.strip().split()
            items.append((name, int(prize), float(weight)))

    n = len(items)

    # Create LP problem
    knapsack_problem = LpProblem("KnapsackProblem", LpMaximize)

    # Binary decision variables
    x = [LpVariable(f"x_{i}", cat="Binary") for i in range(n)]

    # Objective function: maximize total prize
    knapsack_problem += lpSum(items[i][1] * x[i] for i in range(n)), "Total_Prize"

    # Constraint: total weight must not exceed max_weight
    knapsack_problem += lpSum(items[i][2] * x[i] for i in range(n)) <= max_weight, "Total_Weight"

    # Solve the problem
    knapsack_problem.solve()

    # Display results
    selected_items = [items[i] for i in range(n) if x[i].value() == 1]
    
    total_prize = sum(item[1] for item in selected_items)
    total_weight = sum(item[2] for item in selected_items)

    print("\033[1mSelected items:")
    for item in selected_items:
        print(f"{item[0]} - Prize: {item[1]}, Weight: {item[2]}")

    print("\nTotal Prize:", total_prize)
    print("Total Weight:", total_weight, '\033[0m')

if __name__ == "__main__":
    input_file = "k1.txt"  # Change this to the path of your input file
    max_weight = 10  # Change this to your predefined value W

    knapsack_solver(input_file, max_weight)
```

## Part 1 Knapsack problem with dependencies

Using python-mip or any other MIP solver solve a variant of the knapsack problem. Here, some items cannot be taken until their dependencies are satisfied. Dependenicies are written in a separate file. All these dependencies are of the form (a | b | c | …) -> d. Here is an example of a file containing dependencies:    

Cryptokey Flashcard   
Flashcard Laptop   
Battery Charger Laptop   
The lines in the file should be treated as Cryptokey -> Flashcard, Flashcard -> Laptop, (Battery OR Charger) -> Laptop.

```python
from pulp import LpMaximize, LpProblem, LpVariable, lpSum, LpInteger

def solve_knapsack_with_dependencies(items, dependencies, max_weight):
    # Create a linear programming problem
    knapsack_problem = LpProblem("KnapsackProblem", LpMaximize)

    # Create binary variables for each item
    item_vars = LpVariable.dicts("Item", items, 0, 1, LpInteger)

    # Objective function: maximize the total prize
    knapsack_problem += lpSum(item_vars[item] * items[item]["prize"] for item in items)

    # Constraint: total weight cannot exceed max_weight
    knapsack_problem += lpSum(item_vars[item] * items[item]["weight"] for item in items) <= max_weight

    # Add dependencies constraints
    check = []
    for dep in dependencies:
        knapsack_problem += lpSum(item_vars[item] for item in dep) >= 1
        for i in dep:
            check.append(i)

    # Solve the problem
    knapsack_problem.solve()

    # Extract the selected items
    selected_items = [item for item in items if item_vars[item].value() == 1 and item in check]

    # Return the result
    return {
        "ObjectiveValue": lpSum(item_vars[item] * items[item]["prize"] for item in items).value(),
        "SelectedItems": selected_items
    }


def read_input_file(file_path):
    items = {}
    with open(file_path, "r") as file:
        for line in file:
            name, prize, weight = line.strip().split()
            items[name] = {"prize": float(prize), "weight": float(weight)}
    return items

def read_dependencies_file(file_path):
    dependencies = []
    with open(file_path, "r") as file:
        for line in file:
            dep = line.strip().split()
            dependencies.append(dep)
    return dependencies

if __name__ == "__main__":
    input_file_path = "k1.txt"  # Change to your input file path
    dependencies_file_path = "dependencies.txt"  # Change to your dependencies file path
    max_weight = 10.0  # Change to your maximum weight

    items = read_input_file(input_file_path)
    dependencies = read_dependencies_file(dependencies_file_path)

    result = solve_knapsack_with_dependencies(items, dependencies, max_weight)

    print("\033[1mObjective Value:", result["ObjectiveValue"])
    print("Selected Items:")
    for item in result["SelectedItems"]:
        print(item)
    print('\033[0m')
```

## Part 2 Knapsack problem with oversize items (5 points)

Using python-mip or other MIP solver to solve a variant of the knapsack problem.     
Here, every item does have volume (as the third column in an input file

Laptop  10  0.5 0.2   
Flashcard   5   0.01 0.02   
Cryptokey 1 0.01 0.02   
Lamp 1 5.0 1.0    
Battery 1 0.05 0.06   
Charger 0 0.01 0.06

Moreover, some of the items are oversized meaning that if being taken they would not allow you to take items with large volume. Oversized items are listed in a separate file with the corresponding limits on the volume of items:

Laptop 0.05   
Lamp 0.01

In this case, if the laptop is taken no items with a volume more than 0.05 can be taken.

```python
from pulp import LpProblem, LpVariable, lpSum, LpMaximize

def read_values_weights(filename):
    values = {}
    weights = {}
    volumes = {}
    oversized_limits = {}
    with open(filename, 'r') as file:
        for line in file:
            line = line.strip().split()
            item = line[0]
            value = int(line[1])
            weight = float(line[2])
            volume = float(line[3])
            values[item] = value
            weights[item] = weight
            volumes[item] = volume

    return values, weights, volumes

def read_oversized_limits(filename):
    oversized_limits = {}
    with open(filename, 'r') as file:
        for line in file:
            line = line.strip().split()
            item = line[0]
            limit = float(line[1])
            oversized_limits[item] = limit
    return oversized_limits

def solve_knapsack_with_oversized_items(values, weights, volumes, oversized_limits, capacity):
    model = LpProblem(name="Knapsack_With_Oversized_Items", sense=LpMaximize)

    # Create binary variables for each item
    item_vars = LpVariable.dicts("Item", values.keys(), 0, 1, cat='Binary')

    # Objective function
    model += lpSum([value * item_vars[item] for item, value in values.items()])

    # Capacity constraint
    model += lpSum([weight * item_vars[item] for item, weight in weights.items()]) <= capacity

    # Add oversized constraints
    for item, limit in oversized_limits.items():
        model += volumes[item] * item_vars[item] <= limit

    # Solve the problem
    model.solve()

    # Extract the selected items
    selected_items = [item for item, var in item_vars.items() if var.value() == 1 and weights[item] <= oversized_limits.get(item, float('inf'))]
    total_prize = sum(values[item] for item in selected_items)
    total_weight = sum(weights[item] for item in selected_items)
    print( "\033[1m\nTotal Prize:", total_prize)
    print("\033[1mTotal Weight:", total_weight)
    for item in selected_items:
        print(f"{item} - Prize: {values[item]}, Weight: {weights[item]}")
    print('\033[0m')
    return selected_items

# Example usage
values_weights_file = 'k2.txt'
oversized_limits_file = 'c.txt'

values, weights, volumes = read_values_weights(values_weights_file)
oversized_limits = read_oversized_limits(oversized_limits_file)
# Define the capacity of the knapsack
capacity = 10

selected_items = solve_knapsack_with_oversized_items(values, weights, volumes, oversized_limits, capacity)


```
